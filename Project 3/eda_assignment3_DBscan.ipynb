**DATA PREPROCESSING**

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
df.info()
df.head()
```

```python
le = LabelEncoder()
df['Education'] = le.fit_transform(df['Education'])
df['Marital_Status'] = le.fit_transform(df['Marital_Status'])
df = df.drop(['Dt_Customer'],axis=1)
```

```python
#first we label encode, then we scale the data, then we apply PCA
pipe = Pipeline([('scaler', StandardScaler()),
        ('pca', PCA())])

#we will use the pipeline to transform the data
pipe.fit(df)
```

```python

sum = np.cumsum(pipe['pca'].explained_variance_ratio_)

data = {
    'Number of Components': np.arange(1, len(pipe['pca'].explained_variance_ratio_) + 1),
    'Cumulative Explained Variance Ratio': sum
}

# Create a Pandas DataFrame from the data
cumsum_df = pd.DataFrame(data)

# Display the table
print(cumsum_df)
```

**AS WE CAN SEE 22 COMPONENTS CAPTURE ALMOST 95% OF VARIANCE**

```python
df.info()
df.head()
```

```python
#first we label encode, then we scale the data, then we apply PCA
pipe = Pipeline([('scaler', StandardScaler()),
        ('pca', PCA(n_components=20))])

#we will use the pipeline to transform the data
df_transformed = pipe.fit_transform(df)
```

```python
df_transformed.shape
df_transformed=pd.DataFrame(df_transformed)
df_transformed.info()   
df_transformed.head()
```

```python
sns.scatterplot(x=df_transformed[0],y=df_transformed[1])
```

**RUNNING DBSCAN**

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

dbscan=DBSCAN(eps=0.1,min_samples=5)
```

```python
dbscan=DBSCAN(eps=1,min_samples=5)
dbscan.fit(df_transformed)
labels = pd.DataFrame(dbscan.labels_,columns=['cluster'])
labels.value_counts()
```

**ANALYZING CLUSTER**

```python
df_transformed = pd.concat([df_transformed,labels],axis=1)
```

```python
df_transformed.head()
```

```python
cluster_means = df_transformed.groupby('cluster').mean()
cluster_means
```

```python
#calculate silhouette score
silhouette_score(df_transformed,dbscan.labels_)
```

**overall DBSCAN performs really badly on both transformed and non_transformed data**
